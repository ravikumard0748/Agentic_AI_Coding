```python
from pydantic import BaseModel , Field

class Movie(BaseModel):
    name : str = Field(descriptions = "Name of the Movie")
    year : int = Field(descriptions = "release Year of the movie")
    director : str = Field(descrptions = "The director of the movie")
    rating : float = Field(descriptions = "Rating of the movie out of 10")
    
```

```python
from pydantic import Field , BaseModel

class Actor(BaseModel):
    name : str = Field(descriptions = "Name of the Actor")
    role : str = Field(descriptions = "Role of the actor in the movie")


class Movie(BaseModel):
    name : str = Field(descriptions = "Name of the Movie")
    year : int = Field(descriptions = "Release Year of the Movie")
    cast : list[Actor]
    genre : list[str]
    budget : float 
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:llama-3.3-70b-versatile")
```

```python
model = model.with_structured_output(Movie)
model.invoke("Give the details about the movie navagraha")
```
