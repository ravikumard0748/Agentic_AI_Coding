```python
from langchain.messages import AIMessage , HumanMessage , SystemMessage
from langchain.chat_models import init_chat_model

model = init_chat_model("groq:llama-3.3-70b-versatile")

```

```python
messages = [
    SystemMessage("You are excellent Mathamatician with having good knowledge all the type of mathametical problems and you need to provide the answer for all the question that are asked by the User"),
    HumanMessage("What is the output of the cos theta / sin theta"),
    AIMessage("To answer for the question we need to look back the properties of this trignometry terms by looking that we can answer them. The answer for the  cos theta/sin theta is must be the cot theta"),
    HumanMessage("Give me only the answer for the message i don't want any other extra things")
]
```

```python
resp = model.invoke(messages)
```
