```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:llama-3.3-70b-versatile")
```

```python
for chunk in model.stream("How the Explainable AI works"):
    print(chunk.content ,end = "|", flush = True)
```

```python
responses = model.batch(
    ["What is XAI",
     "What is good comapanies to work as ai engineer by considering all ",
     "what is the most affecting job by the ai and how the frontend developer affected by this "]
)
```

```python
for resp in responses:
    print(resp.content)
```

```python
from langchain.tools import tool

@tool
def get_weather(city : str) -> str:
    """Get the Weather from the city"""
    return f"The weather in {city} is sunny"

```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3.6-27b")
model_with_tool = model.bind_tools([get_weather])
```

```python
response = model_with_tool.invoke("What is the Weather in the Boston now")
```
